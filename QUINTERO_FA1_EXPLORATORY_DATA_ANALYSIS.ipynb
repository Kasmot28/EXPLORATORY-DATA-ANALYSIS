<h3>Choose any two of the columns of the dataset and make a Q-Q plot comparing the distributions. What does the Q-Q plot tell you about similarities or differences between the distributions of the values in the two columns?</h3>

```python
import scipy.stats as stats
column1 = np.array(df['NKp30'])
column2 = np.array(df['KIR3DL1'])

plt.figure(figsize=(10,6))
stats.probplot(column1, dist="norm", plot=plt)
stats.probplot(column2, dist="norm", plot=plt)
plt.title('Q-Q Plot of NKp30 and KIR3DL1')
plt.show()
```

Both of the columns have data points (blue dots) that didn't follow the reference line. Saying that both of the columns do not follow the normal distribution. We can see it also at the lower quantiles as it has a deviation from the reference line. Aside from the deviation, both columns show that there are divergences in the middle quantiles, stating the differences in the spread of skewness. And lastly, the behavior of the tail: we can see that the upper quantiles slightly align to our reference line but diverge at the extremes. And in lower quantiles, we can see that there is a potential strongly skewed distribution.
